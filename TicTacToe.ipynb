```python
import numpy as np #Necessary packages
import random
from IPython.display import clear_output
```

```python
IDs = {"guest":[0,0,0]} #Account storage and default account
```

```python
#Fundamentals
def namereg(player, marker, oppmarker, difficulty): #Register new account
    player = input("Please enter your name: ")
    IDs.update({"{php}".format(php = player):[0,0,0]})
    startmenu(player, marker, oppmarker, difficulty)
def loadname(player, marker, oppmarker, difficulty): #Load existing name
    player = input("Please enter your name: ")
    startmenu(player, marker, oppmarker, difficulty)
def updatewin(player): #Updates player statistics
    IDs[player][0] += 1
def updateloss(player): 
    IDs[player][1] += 1
def updatetie(player): 
    IDs[player][2] += 1
def returnname(player): #Returns player statistics
    return player
def returnwins(player): 
    return IDs[player][0]
def returnlosses(player):
    return IDs[player][1]
def returnties(player):
    return IDs[player][2]
def print_all(player): #Scoreboard
    print("{phs}'s record:".format(phs = player))
    print("wins: {phw}".format(phw = IDs[player][0]), "losses: {phl}".format(phl = IDs[player][1]), "ties: {pht}".format(pht = IDs[player][2]))
```

```python
#Start menu and default modes
def startmenu(player = "guest", marker = "O", oppmarker = "X", difficulty = "easy"):
    clear_output(wait = False)
    print("\n", "Welcome to Tic-Tac-Toe!", "\n", "\n", "Select options:") #List of available options
    print(["Player: {phn}".format(phn = player), "Skin: {phm}".format(phm = marker), "Difficulty: {phd}".format(phd = difficulty)])
    print("Wins: {phw}".format(phw = returnwins(player)), "Losses: {phl}".format(phl = returnlosses(player)), "Ties: {pht}".format(pht = returnties(player)))
    print("\n", "COMMAND LIST")
    print("Type 1 to register name")
    print("Type 2 to sign in")
    print("Type 3 to select circles or crosses")
    print("Type 4 to select difficulty")
    print("Type 5 to start game!")
    print("Type 6 to enter two-player mode!")
    print("Type 7 to quit game")
    x = input("Type selection now: ")
    if x == "1":
        namereg(player, marker, oppmarker, difficulty)
    elif x == "2":
        loadname(player, marker, oppmarker, difficulty)
    elif x == "3":
        circlecross(player, marker, oppmarker, difficulty)
    elif x == "4":
        selectdiff(player, marker, oppmarker, difficulty)
    elif x == "5":
        turnselect(player, marker, oppmarker, difficulty)
    elif x == "6":
        namereg2P(player, marker, oppmarker)
    elif x == "7":
        print("\n", "Exited Tic-Tac-Toe.")
        return
    else:
        startmenu(player, marker, oppmarker, difficulty)
```

```python
#Selects circles or crosses
def circlecross(player, marker, oppmarker, difficulty): 
    x = input("Type 0 to select circles(O). Type 1 to select crosses(X): ")
    if x == "0":
        startmenu(player, marker = "O", oppmarker = "X", difficulty = difficulty)
    elif x == "1":
        startmenu(player, marker = "X", oppmarker = "O", difficulty = difficulty)
    else:
        print("Please select valid answer.")
        circlecross(player, marker, oppmarker, difficulty)
```

```python
#Selects difficulty
def selectdiff(player, marker, oppmarker, difficulty): 
    x = input("Type 0 to select easy difficulty. Type 1 to select normal difficulty. Type 2 to select hard difficulty: ")
    if x == "0":
        startmenu(player, marker, oppmarker, difficulty = "easy")
    elif x == "1":
        startmenu(player, marker, oppmarker, difficulty = "normal")
    elif x == "2":
        startmenu(player, marker, oppmarker, difficulty = "hard")
    else:
        print("Please select valid answer.")
        selectdiff(player, marker, oppmarker, difficulty)
```

```python
#Selects turn order
def turnselect(player, marker, oppmarker, difficulty): 
    clear_output(wait = False)
    coin = np.random.rand()
    order = input("Type 0 to go first. Type 1 to go second. Type 2 to flip a coin. Type 3 to return to main menu: ")
    if order == "0" or (order == "2" and coin >= .5):
        turnorder = "pf" #Player goes first
        vscomp(player, marker, oppmarker, difficulty, turnorder)
    elif order == "1" or (order == "2" and coin < .5):
        turnorder = "cf" #Computer goes first
        vscomp(player, marker, oppmarker, difficulty, turnorder)
    elif order == "3":
        startmenu(player)
    else:
        print("Please select valid answer.")
        turnselect(player, marker, oppmarker, difficulty)
```

```python
#Easy computer's decision-making process
def easyaiturn(boarddisp, commlistuse, oppmarker): 
    print("Computer's Turn")
    x = random.choice(commlistuse) #Random entry chosen from list of all possible actions
    boarddisp[tuple(x)] = oppmarker #Entry marked with X/O
    commlistuse.remove(x) #Entry removed from list of actions, so it cannot be selected again
    print(boarddisp) #New board printed
```

```python
#Hard computer's decision-making process
def hardaiturn(boarddisp, commlistuse, marker, oppmarker): 
    print("Computer's Turn")
    if boarddisp[1,1] == "_": #Hard AI must select center spot if available
        boarddisp[1,1] = oppmarker
        commlistuse.remove([1,1])
        print(boarddisp)
        return
    if winseeker(boarddisp, oppmarker) == "w": #Checks for and acts on winning situations
        print("TicTacToe Found!")
        print(boarddisp)
        return
    if lossavoid(boarddisp, marker, oppmarker, commlistuse) == "l": #Checks for and guards against losing situations
        print("Imminent Loss Detected!")
        print(boarddisp)
        return
    x = random.choice(commlistuse) # Random entry chosen from list of all possible actions if no other conditions fulfilled
    boarddisp[tuple(x)] = oppmarker # Random entry marked with X/O
    commlistuse.remove(x) # Entry removed from list of actions
    print(boarddisp) # New board printed
```

```python
#Player's game interface
def playerturn1(boarddisp, commlistuse, marker, player): 
    PC = {"1":(0,0), "2":(0,1), "3":(0,2), "4":(1,0), "5":(1,1), "6":(1,2), "7":(2,0), "8":(2,1), "9":(2,2)}
    print("{phn}'s Turn".format(phn = player))
    print("\n", "Input a command to mark the square you want.") #List of all possible commands:
    print("1: Upper-Left, 2: Upper-Center, 3: Upper-Right")
    print("4: Middle-Left, 5: Middle-Center, 6: Middle-Right")
    print("7: Bottom-Left, 8: Bottom-Center, 9: Bottom-Right")
    print("0: Return to main menu")
    y = input("Please type a command: ")
    if y == "0":
        startmenu(player)
        print("Returned to main menu.")
    else:
        if boarddisp[PC[y]] != "_": #If space is filled, repeats function
            print("That space is filled! Please select another.")
            playerturn1(boarddisp, commlistuse, marker, player)
        elif boarddisp[PC[y]] == "_": #If space is not filled;
            boarddisp[PC[y]] = marker #Places X/O at said space;
            commlistuse.remove(list(PC[y])) #Removes said space from enemy AI's possibile choices
            print(boarddisp) #Prints new board
```

```python
#Post-tie options, reached if turn count exceeds nine
def tie(player, marker, oppmarker, difficulty): 
    updatetie(player) #updates, displays player's ties
    print_all(player)
    x = input("Tie reached! Reset board? Type 0 if yes. Type 1 to return to main menu: ")
    if x == "0":
        clear_output(wait = False)
        turnselect(player, marker, oppmarker, difficulty)
    elif x == "1":
        clear_output(wait = False)
        startmenu(player)
```

```python
#Postgame options, reached if player/computer wins
def postgame(player, marker, oppmarker, difficulty): 
    x = input("Reset board? Type 0 if yes. Type 1 to return to main menu: ")
    if x == "0":
        clear_output(wait = False)
        turnselect(player, marker, oppmarker, difficulty)
    elif x == "1":
        clear_output(wait = False)
        startmenu(player)
```

```python
#Comprehensive engine for player vs computer matches
def vscomp(player, marker, oppmarker, difficulty, turnorder): 
    boarddisp = np.array([["_", "_", "_"], ["_", "_", "_"], ["_", "_", "_"]]) #Fundamental assets; blank board
    commlistuse = [[0, 0], [0, 1], [0, 2], [1, 0], [1, 1], [1, 2], [2, 0], [2, 1], [2, 2]] #Computer's commands
    turn = 1
    print(boarddisp)
    if turnorder == "pf": #If player goes first:
        while turn <= 10: #gameloop starts
            print("\n", "Turns: {pht}".format(pht = turn)) #Turn counter
            playerturn1(boarddisp, commlistuse, marker, player) #Player interface
            if wincheck(boarddisp, marker) == "PW": #Checks for winning combinations
                print("Player Wins!")
                updatewin(player) #Updates, reports scores if winning combination achieved
                print_all(player)
                break #Breaks gameloop
            else:
                pass
            turn += 1 #Advances turn counter
            if turn == 10: #Activates tie()
                tie(player, marker, oppmarker, difficulty)
            else:
                pass
            print("\n", "Turns: {pht}".format(pht = turn))
            coin = np.random.rand() #Decision tree for computer behavior based on difficulty starts
            if difficulty == "easy" or (difficulty == "normal" and coin > .5): 
                easyaiturn(boarddisp, commlistuse, oppmarker)
            elif difficulty == "hard" or (difficulty == "normal" and coin <= .5):
                hardaiturn(boarddisp, commlistuse, marker, oppmarker)
            if losecheck(boarddisp, oppmarker) == "CW": #Checks for losing combinations(winning for computer)
                print("Computer Wins!")
                updateloss(player) #Updates, reports scores if losing combination achieved
                print_all(player)
                break 
            else:
                pass
            turn += 1
    elif turnorder == "cf": #If computer goes first; reverses prior order of events
        while turn <= 10:
            print("\n", "Turns: {pht}".format(pht = turn))
            coin = np.random.rand()
            if difficulty == "easy" or (difficulty == "normal" and coin > .5):
                easyaiturn(boarddisp, commlistuse, oppmarker)
            elif difficulty == "hard" or (difficulty == "normal" and coin <= .5):
                hardaiturn(boarddisp, commlistuse, marker, oppmarker)
            if losecheck(boarddisp, oppmarker) == "CW":
                print("Computer Wins!")
                updateloss(player)
                print_all(player)
                break 
            else:
                pass
            turn += 1
            if turn == 10:
                tie(player, marker, oppmarker, difficulty)
            else:
                pass
            print("\n", "Turns: {pht}".format(pht = turn))
            playerturn1(boarddisp, commlistuse, marker, player) #User Interface
            if wincheck(boarddisp, marker) == "PW":
                print("Player Wins!")
                updatewin(player)
                print_all(player)
                break
            else:
                pass
            turn += 1
    postgame(player, marker, oppmarker, difficulty) #activates postgame() upon gameloop end
```

```python
#Checks for all 8 losing conditions, reports if computer/Player 2 wins
def losecheck(boarddisp, oppmarker):
    print("Running losecheck...")
    if boarddisp[0, 0] == boarddisp[0, 1] == boarddisp[0, 2] == oppmarker:
        return "CW"
    elif boarddisp[1, 0] == boarddisp[1, 1] == boarddisp[1, 2] == oppmarker:
        return "CW"
    elif boarddisp[2, 0] == boarddisp[2, 1] == boarddisp[2, 2] == oppmarker:
        return "CW"
    elif boarddisp[0, 0] == boarddisp[1, 0] == boarddisp[2, 0] == oppmarker:
        return "CW"
    elif boarddisp[0, 1] == boarddisp[1, 1] == boarddisp[2, 1] == oppmarker:
        return "CW"
    elif boarddisp[0, 2] == boarddisp[1, 2] == boarddisp[2, 2] == oppmarker:
        return "CW"
    elif boarddisp[0, 0] == boarddisp[1, 1] == boarddisp[2, 2] == oppmarker:
        return "CW"
    elif boarddisp[0, 2] == boarddisp[1, 1] == boarddisp[2, 0] == oppmarker:
        return "CW"
    else:
        print("OK")
```

```python
#Additional directions for Hard AI to avoid losing situations; constantly analyzes for losing situations and
#blocks players from getting a TicTacToe. All eight losing conditions are monitored for.
def lossavoid(boarddisp, marker, oppmarker, commlistuse):
    print("Running lossavoid...")
    lossimminent = marker + marker #lossimminent will equal "XX" or "OO" depending on marker player selected.
    if ((boarddisp[0, 0] + boarddisp[0, 1] + boarddisp[0, 2]).replace("_", "")) == lossimminent:
        lossdetector([0,0], [0,1], [0,2], boarddisp, commlistuse, oppmarker)
        return "l"
    #If the sum of any two values for these three positions(in this case the first row) is "lossimminent,"
    #lossdetector() activates.
    elif ((boarddisp[1, 0] + boarddisp[1, 1] + boarddisp[1, 2]).replace("_", "")) == lossimminent:
        lossdetector([1,0], [1,1], [1,2], boarddisp, commlistuse, oppmarker)
        return "l"
    elif ((boarddisp[2, 0] + boarddisp[2, 1] + boarddisp[2, 2]).replace("_", "")) == lossimminent:
        lossdetector([2,0], [2,1], [2,2], boarddisp, commlistuse, oppmarker)
        return "l"
    elif ((boarddisp[0, 0] + boarddisp[1, 0] + boarddisp[2, 0]).replace("_", "")) == lossimminent:
        lossdetector([0,0], [1,0], [2,0], boarddisp, commlistuse, oppmarker)
        return "l"
    elif ((boarddisp[0, 1] + boarddisp[1, 1] + boarddisp[2, 1]).replace("_", "")) == lossimminent:
        lossdetector([0,1], [1,1], [2,1], boarddisp, commlistuse, oppmarker)
        return "l"
    elif ((boarddisp[0, 2] + boarddisp[1, 2] + boarddisp[2, 2]).replace("_", "")) == lossimminent:
        lossdetector([0,2], [1,2], [2,2], boarddisp, commlistuse, oppmarker)
        return "l"
    elif ((boarddisp[0, 0] + boarddisp[1, 1] + boarddisp[2, 2]).replace("_", "")) == lossimminent:
        lossdetector([0,0], [1,1], [2,2], boarddisp, commlistuse, oppmarker)
        return "l"
    elif ((boarddisp[0, 2] + boarddisp[1, 1] + boarddisp[2, 0]).replace("_", "")) == lossimminent:
        lossdetector([0,2], [1,1], [2,0], boarddisp, commlistuse, oppmarker)
        return "l"
    else:
        print("OK")
```

```python
#Necessary function for lossavoid() to choose where to mark to avoid player TicTacToe.
#Parses through the given board coordinates for blank spaces and fills them.
def lossdetector(a, b, c, boarddisp, commlistuse, oppmarker):
    if boarddisp[tuple(a)] == "_":
        boarddisp[tuple(a)] = oppmarker
        commlistuse.remove(a)
    elif boarddisp[tuple(b)] == "_":
        boarddisp[tuple(b)] = oppmarker
        commlistuse.remove(b)
    elif boarddisp[tuple(c)] == "_":
        boarddisp[tuple(c)] = oppmarker
        commlistuse.remove(c)
    else:
        return
```

```python
#Checks for all 8 winning conditions, reports if Player 1 (human) wins
def wincheck(boarddisp, marker):
    print("Running wincheck...")
    if boarddisp[0, 0] == boarddisp[0, 1] == boarddisp[0, 2] == marker:
        return "PW"
    elif boarddisp[1, 0] == boarddisp[1, 1] == boarddisp[1, 2] == marker:
        return "PW"
    elif boarddisp[2, 0] == boarddisp[2, 1] == boarddisp[2, 2] == marker:
        return "PW"
    elif boarddisp[0, 0] == boarddisp[1, 0] == boarddisp[2, 0] == marker:
        return "PW"
    elif boarddisp[0, 1] == boarddisp[1, 1] == boarddisp[2, 1] == marker:
        return "PW"
    elif boarddisp[0, 2] == boarddisp[1, 2] == boarddisp[2, 2] == marker:
        return "PW"
    elif boarddisp[0, 0] == boarddisp[1, 1] == boarddisp[2, 2] == marker:
        return "PW"
    elif boarddisp[0, 2] == boarddisp[1, 1] == boarddisp[2, 0] == marker:
        return "PW"
    else:
        print("OK")
```

```python
#Additional directions for Hard AI to pursue winning opportunities if available; checks for instances of where
#TicTacToe can be achieved and goes for them
def winseeker(boarddisp, oppmarker):
    print("Running winseeker...")
    winimminent = oppmarker + oppmarker
    if ((boarddisp[0, 0] + boarddisp[0, 1] + boarddisp[0, 2]).replace("_", "")) == winimminent:
        boarddisp[0, 0] = boarddisp[0, 1] = boarddisp[0, 2] = oppmarker
        return "w"
    #If two spaces in the first row contain "oppmarker" and the other is blank, that blank space is taken and the
    #computer wins. Takes priority over lossavoid().
    elif ((boarddisp[1, 0] + boarddisp[1, 1] + boarddisp[1, 2]).replace("_", "")) == winimminent:
        boarddisp[1, 0] = boarddisp[1, 1] = boarddisp[1, 2] = oppmarker
        return "w"
    elif ((boarddisp[2, 0] + boarddisp[2, 1] + boarddisp[2, 2]).replace("_", "")) == winimminent:
        boarddisp[2, 0] = boarddisp[2, 1] = boarddisp[2, 2] = oppmarker
        return "w"
    elif ((boarddisp[0, 0] + boarddisp[1, 0] + boarddisp[2, 0]).replace("_", "")) == winimminent:
        boarddisp[0, 0] = boarddisp[1, 0] = boarddisp[2, 0] = oppmarker
        return "w"
    elif ((boarddisp[0, 1] + boarddisp[1, 1] + boarddisp[2, 1]).replace("_", "")) == winimminent:
        boarddisp[0, 1] = boarddisp[1, 1] = boarddisp[2, 1] = oppmarker
        return "w"
    elif ((boarddisp[0, 2] + boarddisp[1, 2] + boarddisp[2, 2]).replace("_", "")) == winimminent:
        boarddisp[0, 2] = boarddisp[1, 2] = boarddisp[2, 2] = oppmarker
        return "w"
    elif ((boarddisp[0, 0] + boarddisp[1, 1] + boarddisp[2, 2]).replace("_", "")) == winimminent:
        boarddisp[0, 0] = boarddisp[1, 1] = boarddisp[2, 2] = oppmarker
        return "w"
    elif ((boarddisp[0, 2] + boarddisp[1, 1] + boarddisp[2, 0]).replace("_", "")) == winimminent:
        boarddisp[0, 2] = boarddisp[1, 1] = boarddisp[2, 0] = oppmarker
        return "w"
    else:
        print("OK")
```

```python
#Necessary to register/load Player 2 info for two player mode
def namereg2P(player, marker, oppmarker): 
    x = input("Are you a new or returning player? Type 0 for new. Type 1 for returning: ")
    if x == "0": #Identical to namereg().
        player2 = input("Please enter your name: ")
        IDs.update({"{php2}".format(php2 = player2):[0,0,0]})
        turnselect2P(player, player2, marker, oppmarker)
    elif x == "1": #Identical to loadname().
        player2 = input("Please enter your name: ")
        turnselect2P(player, player2, marker, oppmarker)
```

```python
#Special turn selector for two player mode
def turnselect2P(player, player2, marker, oppmarker): 
    clear_output(wait = False) #All "2P" functions must share Player 2's data.
    print("{php}".format(php = player), "versus", "{php2}".format(php2 = player2))
    coin = np.random.rand()
    order = input("Type 0 to go first. Type 1 to go second. Type 2 to flip a coin. Type 3 to return to main menu: ")
    if order == "0" or (order == "2" and coin >= .5):
        turnorder = "p1f"
        print("{php} goes first!".format(php = player))
        vscomp2P(player, player2, marker, oppmarker, turnorder)
    elif order == "1" or (order == "2" and coin < .5):
        turnorder = "p2f"
        print("{php2} goes first!".format(php2 = player2))
        vscomp2P(player, player2, marker, oppmarker, turnorder)
    elif order == "3":
        startmenu(player)
    else:
        print("Please select valid answer.")
        turnselect2P(player, player2, marker, oppmarker)
```

```python
#Special engine for two player mode
def vscomp2P(player, player2, marker, oppmarker, turnorder): 
    boarddisp = np.array([["_", "_", "_"], ["_", "_", "_"], ["_", "_", "_"]])
    commlistuse = [[0, 0], [0, 1], [0, 2], [1, 0], [1, 1], [1, 2], [2, 0], [2, 1], [2, 2]]
    turn = 1
    print(boarddisp) #Blank starting board
    if turnorder == "p1f":
        while turn <= 10:
            print("\n", "Turns: {pht}".format(pht = turn))
            playerturn1(boarddisp, commlistuse, marker, player)
            if wincheck(boarddisp, marker) == "PW":
                print("{php} Wins!".format(php = player))
                updatewin(player)
                print_all(player)
                updateloss(player2)
                print_all(player2)
                break
            else:
                pass
            turn += 1
            if turn == 10:
                tie2P(player, player2, marker, oppmarker)
            else:
                pass
            print("\n", "Turns: {pht}".format(pht = turn))
            playerturn2(boarddisp, commlistuse, oppmarker, player2)
            if losecheck(boarddisp, oppmarker) == "CW":
                print("{php2} Wins!".format(php2 = player2))
                updateloss(player)
                print_all(player)
                updatewin(player2)
                print_all(player2)
                break 
            else:
                pass
            turn += 1
    elif turnorder == "p2f":
        while turn <= 10:
            print("\n", "Turns: {pht}".format(pht = turn))
            playerturn2(boarddisp, commlistuse, oppmarker, player2)
            if losecheck(boarddisp, oppmarker) == "CW":
                print("{php2} Wins!".format(php2 = player2))
                updateloss(player)
                print_all(player)
                updatewin(player2)
                print_all(player2)
                break 
            else:
                pass
            turn += 1
            if turn == 10:
                tie2P(player, player2, marker, oppmarker)
            else:
                pass
            print("\n", "Turns: {pht}".format(pht = turn))
            playerturn1(boarddisp, commlistuse, marker, player)
            if wincheck(boarddisp, marker) == "PW":
                print("{php} Wins!".format(php = player))
                updatewin(player)
                print_all(player)
                updateloss(player2)
                print_all(player2)
                break
            else:
                pass
            turn += 1
    x = input("Reset board? Type 0 if yes. Type 1 to return to main menu: ") #Special postgame code for two player mode.
    if x == "0":
        clear_output(wait = False)
        turnselect2P(player, player2, marker, oppmarker) #Redirects to two player turn select.
    elif x == "1":
        clear_output(wait = False)
        startmenu(player)
```

```python
#Special tie function for two player mode
def tie2P(player, player2, marker, oppmarker): 
    updatetie(player) #Updates, prints both player's statistics.
    print_all(player)
    updatetie(player2)
    print_all(player2)
    x = input("Tie reached! Reset board? Type 0 if yes. Type 1 to return to main menu: ")
    if x == "0":
        clear_output(wait = False)
        turnselect2P(player, player2, marker, oppmarker) #Redirects to two player turn select.
    elif x == "1":
        clear_output(wait = False)
        startmenu(player)
```
